```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, p = 13, 0.7
s = np.random.binomial(n, p, 1000)
```

```python
width = 0.1
histmin = np.floor(s)
histmax = np.ceil(s)

plt.hist(s, bins = 100, alpha = 0.5, edgecolor = 'black')
plt.xlabel('sucessess')
plt.ylabel('')

plt.show
plt.savefig('binomial dist.pdf')
```
