```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
cosx = np.cos(x)
sinx = np.sin(x)
```

```python
figure, axes = plt.subplots(1,2)

axes[1].plot(x, sinx)
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].set_title('cos(x)')



axes[0].plot(x, sinx)
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].set_title('sin(x)')


```
